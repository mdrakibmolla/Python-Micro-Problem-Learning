**🖨️ print output – 5 Micro Problems (Python)**

```python
# 1. Simple print

print("Hello World")

# example
# print()-> স্ক্রিনে লেখা দেখায়
# "" --> String
```

```python
# 2. variable print

Name = input("Your Name: ")
print("Hello", Name)

# example
# print("Hello", Name)-> Space দিয়ে join হয়
```

```python
# 3. f-String দিয়ে print (modern way)

name = input("Name: ")
age = int(input("Age: "))
print(f"My name is {name} and I am {age} year old")

# example
# f"" -> Variable embed করা যায়
```

```python
# 4. Separator Use Case

print("Python", "Java", "C++", sep = " - ")

#example
# sep -> আলাদা করার চিহ্ন
```

```python
# 5. End Parameter Use case

print("Hello", end =" Rakib ")
print("World")

# example
# default end = "\n" (new line)
# এখানে " " দিলে same line এ print হয়
```

```python
# 6. Multiple line print

print("Line 1")
print("Line 2")
print("Line 3")

# example
# print default নতুন লাইনে যায় (\n)
```

```python
# 7. New line escape(\n)

print("Hello\nWorld")

# example
# \n-> manual new line
```

```python
# 8. tap space(\t)

print("Name:\tRakib")
print("Age:\t20")

# example
# \t-> tab space দেয় (table style)
```

```python
# 9. Number formatting (2 decimal)

num = float(input("Number: "))
print(f"{num:.2f}")

# example
# {num:.2f} → 2 decimal পর্যন্ত দেখায়
```

```python
# 10. Table style output

name = input("Name: ")
age = input("Age: ")

print("_____INFO_____")
print("Name:", name)
print("Age:", age)
print("__________")

# simple formatting দিয়ে UI এর মতো output বানানো যায়
```

**print output- 5 Micro problem (new)**

```python
# 1. Quotes সহ print

print('He said "Hello"')
print("Hello 'World'")

# example
# single ' ' আর double " " mix করলে quotes দেখানো যায়
```

```python
# 2. Backslash print করা

print("c:\\user\\rakib")

# example
# \\ → একটাকে \ হিসেবে দেখায়
```

```python
# 3. Expression print

a = int(input("a: "))
b = int(input("b: "))

print("Sum =", a + b)

# example
# print() এর ভিতরে calculation করা যায়
```

```python
# 4. Multiple Variable print(f-string)

name = input("Name: ")
city = input("City: ")

print(f"{name} lives in {city}")
```

```python
# 5. Same line input/output style

num = input("Enter number: ")

print("you entered", num)
```

```python
# 6. String input নিয়ে print

name = input("Your Name: ")
print(name)

# example
# input() default string নেয়
```

```python
# 7. Integer input নিয়ে যোগ

a = int(input("a: "))
b = int(input("b: "))

print(a+b)

# example
# int() → string কে integer বানায়
```

```python
# 8. Float input নিয়ে গুণ

a = float(input())
b = float(input())

print(a*b)
```

```python
# 9. Multiple input (এক লাইনে)

a, b = map(int, input("Two number: ").split())

print(a + b)

# example
# split() → string ভাগ করে
# map(int, ...) → int বানায়
```

```python
# 10. List input

nums = list(map(int, input("number: ").split()))

print(nums)
```
